```python
import numpy as np
from numpy import linalg as la
```

```python
xi = np.array([[0.0115,0.012,0.013,0.026,0.034,0.04,0.084,0.092]])
yi = np.array([[50.5, 50.2,48.5,35,38,28,15,20.5]])
```

```python
coef = np.array([[-1], [0]])
alpha = 0.001
error = np.sum((yi-coef[0,0]*np.exp(coef[1,0]*xi))**2)
```

```python
coef = coef - alpha*np.array([[-2*np.sum((yi-coef[0,0]*np.exp(coef[1,0]*xi))*(np.exp(coef[1,0]*xi)))],
                      [-2*np.sum((yi-coef[0,0]*np.exp(coef[1,0]*xi))*(coef[0,0]*xi*np.exp(coef[1,0]*xi)))]])
```

```python
print(error)
print(np.sum((yi-coef[0,0]*np.exp(coef[1,0]*xi))**2))
```
