```python
import pandas as pd
import numpy as np


pelis = pd.read_csv('ratings_example.csv')
pelis
```

```python
pelis_pivoted = pelis.pivot(index='userId',columns='movieId',values='rating').fillna(0)
pelis_pivoted
```

```python
M = pelis_pivoted.to_numpy()
print(M,'\n', M.shape)
```

```python
#Vamos a analizar usuarios porque son muchos menos que pelÃ­culas

from numpy import linalg as la
#Usuarios normalizados por filas MN
MN= (M.T/la.norm(M.T,axis=0)).T
print(MN)
```

```python
#Matriz de similitud entre usuarios
S=MN@MN.T
print(S)
```

```python
#recomendaciÃ³n para el usuario 1
print(S[0:1,:]@M)
rounded_result = np.round(S[0:1,:] @ M, 3)
print(rounded_result)
```

```python
# funciÃ³n con el que le dices el usuario y coger el vectro y obtener el vector de recomendaciÃ³n
def recomendacion(usuario):
    x= np.round(S[usuario-1:usuario,:] @ M, 3)
    x[M[usuario-1:usuario,:]!=0] = 0
    return pelis_pivoted.columns[x.argmax()]
```

```python
recomendacion(586)
```
