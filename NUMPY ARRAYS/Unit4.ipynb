```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.array(['Q1','Q2','Q3','Q4'])
y = np.array([234,254,144,357])
```

```python
x,y
```

```python
plt.bar(x,y,color='blue')
plt.title('BAR CHART')
plt.show()
```

```python
plt.hist(x, bins=20)
plt.title('Histogram of x')
plt.show()
```

```python
plt.boxplot(y)
plt.show()
```

```python
x = np.random.randn(10000)
plt.hist(x, bins=200, color='red', density=True, alpha=0.3)
plt.title('Histogram of x')
plt.show()
```

```python
x = np.random.randn(10000)*10+3
y = np.random.randn(10000)*10+5
z = np.random.randn(10000)*10+1

plt.boxplot([x,y,z],0)
plt.title('Boxplot of x, y, z')
plt.show()
```

```python

plt.boxplot([x,y,z],0,vert=False)
plt.title('Boxplot of x, y, z')
plt.show()
```

```python
x = np.linspace(0,10,100) # 100 numbers from 0 to 10
y = np.sin(x)
```

```python
plt.plot(x,y, color='orange', linewidth=7)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.show()
```
