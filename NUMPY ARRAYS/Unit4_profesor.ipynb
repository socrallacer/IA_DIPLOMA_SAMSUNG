```python
# slide 457

x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(figsize=(5,2), dpi=100)
ax.plot(x, y, color='orange', linestyle='--', label='Sin')
ax.plot(x, z, color='purple', linestyle='-', label='Cos')
ax.set_xlabel('X')
ax.set_ylabel('Sin')
ax.set_title('LINE PLOT')
ax.legend(loc=0)
plt.show()
```

```python
# slide 458

x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(ncols=2, figsize=(8,3), dpi=80)
ax[0].plot(x, y, color='orange', linestyle='--', label='Sin')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_title('SINE')

ax[1].plot(x, z, color='purple', linestyle='-', label='Cos')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].set_title('COSINE')

plt.show()
```

## Pandas visualization

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data_Iris.csv')
df.head()
```

```python
df.shape
```

```python
df.loc[:,'Petal.Length'].plot.hist(color='green', bins=50, alpha=0.3)
plt.show()
```

```python
df['Petal.Length'].plot.hist(color='blue', alpha=0.5, rwidth=0.9)
```

```python
#rwidth es el ancho de las barras, permite percibir espacio entre ellas
df['Petal.Length'].plot.hist(color='blue', alpha=0.5, rwidth=0.95)
```

```python
#Alpha especifica la transferencia de las barras
df[['Sepal.Width','Petal.Length']].plot.hist(bins=20,
                                             color=['red','blue'],
                                             alpha=0.6, rwidth=0.9)
plt.show()                                            
```

```python
frequencies = df.Species.value_counts()
my_counts = list(frequencies.values)
my_labels = list(frequencies.index)
df2 = pd.DataFrame( {'counts':my_counts }, index=my_labels )
df2
```

```python
df2.plot.bar(color='orange', alpha=0.7)
plt.show()
```

```python
df.plot.scatter(x = 'Petal.Length',
                y = 'Petal.Width',
                color = 'blue',
                alpha = 0.5,
                marker = 'o')
```

```python
# slide 469 an array of scatter plots

my_cols_dict = { 'setosa':'red', 
                 'virginica':'green', 
                 'versicolor':'blue' }

my_cols = df.Species.apply(lambda x: my_cols_dict[x])

pd.plotting.scatter_matrix(df, c=my_cols, marker='o', alpha=0.5)
plt.show()
```

## Seaborn visualization library

```python
import pandas as pd
import numpy as np
import os
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dat = sns.load_dataset('mpg')
```

```python
dat.shape
```

```python
dat.head()
```

```python
# Histogram (slide 476)

sns.displot(dat.mpg, kde=False, rug=False, bins=50, color='red')
plt.show()
```

```python
# Histogram + KDE

sns.displot(dat.mpg, kde=True, rug=False, bins=50, color='green')
plt.show()
```

```python
# Histogram + RUG

sns.displot(dat.mpg, kde=False, rug=True, bins=50, color='blue',alpha=0.5)
plt.show()
```

```python
# KDE (kernel density estimation)

sns.kdeplot(dat.mpg, color='blue')
plt.show()
```

```python
# Scatter plot (slide 480)

sns.jointplot(x='weight', y='mpg', 
              data=dat, color='red', kind='scatter')
plt.show()
```

```python
# Scatter plot + regression line

sns.jointplot(x='weight', y='mpg', 
              data=dat, color='orange', kind='reg')
plt.show()
```

```python
# Scatter plot + regression line

sns.lmplot(x='weight', y='mpg', data=dat)
plt.show()
```

```python
# Scatter plot + regression line (using 'hue' argument)

sns.lmplot(x='weight', y='mpg', data=dat, hue='origin')
plt.show()
```

```python
# Scatter plot + regression line (using 'hue' argument)

sns.lmplot(x='weight', y='mpg', data=dat, col='origin', hue='origin')
plt.show()
```

```python
# Scatter plot + regression line (using 'hue' argument)

sns.lmplot(x='weight', y='mpg', data=dat, 
           col='origin', hue='origin',
           markers=['o','v','^'])
plt.show()
```

```python
sns.barplot(x='origin', y='mpg', data=dat)
plt.show()
```

```python
sns.barplot(x='origin', y='mpg', data=dat, estimator=np.median)
plt.show()
```

```python
sns.countplot(x='origin',data=dat,palette='coolwarm')
plt.show()
```

```python
sns.countplot(x='origin',data=dat,hue='cylinders')
plt.show()
```

```python
sns.boxplot(x='mpg', y='origin', data=dat)
```

```python
sns.boxplot(x='origin', y='mpg', data=dat, hue='cylinders')
```

```python
sns.violinplot(x='origin', y='mpg', data=dat)
plt.show()
```

```python
sns.stripplot(x='cylinders',y='horsepower',data=dat, palette='seismic')
plt.show()
```

```python
sns.pairplot(dat)
plt.show()
```
