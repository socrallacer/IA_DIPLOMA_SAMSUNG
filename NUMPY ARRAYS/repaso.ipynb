```python
#paso 1) Obtener los 2 años que más modelos se fabricaron
años_mas = df.model_year.value_counts()[:2].index
años_mas
```

```python
#paso 2) Filtrar los coches fabricados en esos años
df.horsepower[df.model_year.isin(años_mas)].max()
```
